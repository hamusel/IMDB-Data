## Average Rating for director

```python
imdb_data.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

## Most popular movie title according to vote count

```python
imdb_data[imdb_data['Votes'].max() == imdb_data['Votes']]['Title']
```

## Most popular movie title according to Revenue

```python
imdb_data[imdb_data['Revenue (Millions)'].max()==imdb_data['Revenue (Millions)']]['Title']
```
